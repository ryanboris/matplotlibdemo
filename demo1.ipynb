```python
from scipy.stats import sem
```

```python
def summary_stats(series):
    return {
        "mean": series.mean(), 
        "median": series.mean(), 
        "variance": series.var(), 
        "std": series.var(), 
        "SEM": sem(series)
    }
```

```python
tumor_vol_series = study_df['Tumor Volume (mm3)']
```

```python
summary_stats(tumor_vol_series)
```

```python
summary_stats(study_df['Metastatic Sites'])
```

```python
df_merged = pd.merge(left=mouse_df, right=study_df, on="Mouse ID")
```

```python
df_merged.describe()
```

```python
df = df_merged.groupby(['Drug Regimen', 'Timepoint'])['Tumor Volume (mm3)'].agg(['count','mean', 'median', 'sem', 'var', 'std'])
```

```python
df
```

```python
df = df.loc[ :'Zoniferol', :].reset_index()
```

```python
propriva = df[df['Drug Regimen'] == 'Propriva']
```

```python
propriva = propriva.rename(columns=lambda x: x.upper())
```

```python
propriva
```

```python
propriva.drop('DRUG REGIMEN', axis=1, inplace=True)
```

```python
propriva.drop('INDEX', axis=1, inplace=True)
```

```python
propriva
```

```python
fig, ax = plt.subplots(ncols=2, nrows=2, figsize=(12, 8))
plt.tight_layout()
plt.delaxes(ax[1, 0])
plt.delaxes(ax[1, 1])
ax[0, 0].plot(propriva.TIMEPOINT, propriva.COUNT)
ax[0, 1].plot(propriva.TIMEPOINT, propriva.MEAN)
plt.show()
```
